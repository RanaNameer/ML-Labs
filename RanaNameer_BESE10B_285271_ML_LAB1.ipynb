```python
x = file.drop('label', axis=1)
y = file['label']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
class Tree:
  depth = 0
  accuracy = 0
  entropy = 0
  def __init__(self, depth, accuracy, entropy) -> None:
      self.depth = depth
      self.accuracy = accuracy
      self.entropy = entropy
  def getAcc():
    return self.accuracy
  def getDepth():
    return self.depth
  def getEntropy():
    return self.entropy
```

```python
arr = []
for i in range(1, len(x.columns)):
  entropy = DecisionTreeClassifier(criterion='entropy', random_state = 0, max_depth = 0)
  entropy.fit(x_train, y_train)
  prediction = entropy.predict(x_test)
  accuracy = accuracy_score(y_test, prediction)*100
  arr.append(Tree(i, accuracy, entropy))
```

```python
maxObj = max(array, key=attrgetter('accuracy'))

plt.xlabel("Depth")
plt.ylabel("Accuracy")
plt.plot([ar.getDepth() for ar in arr],[ar.getAcc() for ar in arr])
plt.show
```
